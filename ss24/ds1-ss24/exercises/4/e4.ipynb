```python
import pandas as pd

# df = pd.read_csv("tips.csv")
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
print(df.shape)

# Check if each dtype is numeric
data_types = df.dtypes
count_numeric = 0
for col, dtype in data_types.items():
    is_numeric = pd.api.types.is_numeric_dtype(dtype)
    if is_numeric:
        count_numeric += 1
    print(f"{col} is numeric: {is_numeric}")

print(
    f"Numeric columns: {count_numeric}, other: {len(data_types) - count_numeric}"
)
```

```python
# Calculate the mean age for each passenger class
mean_bill = df['total_bill'].mean(numeric_only=True)
print(mean_bill)
```

```python
import seaborn as sns

sns.set_theme() # make it pretty

sns.scatterplot(data=df, x="total_bill", y="tip", hue="tip")
```

```python
# Calculate mean tip by sex
mean_tip_by_sex = df.groupby("sex")["tip"].mean()
mean_tip_male = mean_tip_by_sex["Male"]
mean_tip_female = mean_tip_by_sex["Female"]
print(f"Mean tip by sex:\nMale: {mean_tip_male}\nFemale: {mean_tip_female}")
print(mean_tip_by_sex)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("titanic.csv")

df.info()

# Further cleaning
# valid_sex_entries = ["male", 'female', "mal", "femal", "mle"]
valid_sex_entries = ["male", 'female']
# Convert all entries to lowercase for comparison
df['Sex'] = df['Sex'].astype(str).str.lower()

# Replace invalid entries with NaN
df.loc[~df['Sex'].isin(valid_sex_entries), 'Sex'] = np.nan

# Fill NaN values with 'unknown'
df['Sex'].fillna('unknown', inplace=True)

print(df)
```
